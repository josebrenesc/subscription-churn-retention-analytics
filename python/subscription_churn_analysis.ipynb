# Subscription Churn & Retention Analytics
Analysis of a **synthetic** subscription portfolio covering churn, revenue, usage, cancellations, reactivations, and cohort retention.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if not (ROOT / 'data').exists():
    ROOT = ROOT.parent
DATA = ROOT / 'data' / 'processed'
```

## 1. Load and validate data

```python
subscriptions = pd.read_csv(DATA / 'subscriptions_clean.csv', parse_dates=['start_date', 'end_date'])
payments = pd.read_csv(DATA / 'payments_clean.csv', parse_dates=['payment_date'])
monthly = pd.read_csv(DATA / 'monthly_subscription_kpis.csv')
subscriptions.head()
```

```python
print('Duplicate subscription IDs:', subscriptions['subscription_id'].duplicated().sum())
display(subscriptions.isna().sum().sort_values(ascending=False))
```

## 2. Portfolio KPIs

```python
kpis = {
    'Total subscriptions': len(subscriptions),
    'Active subscriptions': subscriptions['subscription_status'].eq('Active').sum(),
    'Cancelled share': subscriptions['subscription_status'].eq('Cancelled').mean(),
    'Collected revenue': payments['paid_amount_usd'].sum(),
    'Refund rate': payments['refund_amount_usd'].sum() / payments['billed_amount_usd'].sum(),
}
kpis
```

## 3. Monthly churn and revenue

```python
monthly[['month', 'active_subscriptions', 'churned_subscriptions', 'churn_rate', 'collected_revenue_usd']].head()
```

```python
monthly.plot(x='month', y='churn_rate', kind='line', marker='o')
plt.title('Monthly Churn Rate')
plt.ylabel('Churn rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Interpretation
- Churn rate, revenue, usage, payment failures, and support experience should be interpreted together.
- A high cancellation count can reflect a large plan population rather than poor relative retention.
- Synthetic data demonstrates analytical workflow and does not establish causal relationships.
